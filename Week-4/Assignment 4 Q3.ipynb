```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
N = 10000         # Total number of needle drops
d = 1.0           # Distance between the lines
l = 1.0           # Length of the needle (l <= d)

# Initialize counters
crossings = 0
pi_estimates = []

# Seed for reproducibility (optional)
np.random.seed(42)

# Monte Carlo Simulation
for i in range(1, N + 1):
    # Random angle between 0 and pi/2
    theta = np.random.uniform(0, np.pi / 2)
    
    # Distance from the center of the needle to the nearest line
    x = np.random.uniform(0, d / 2)

    # Check if the needle crosses a line
    if (l / 2) * np.sin(theta) >= x:
        crossings += 1
    
    # Estimate π after each drop
    if crossings > 0:
        pi_estimate = (2 * i) / (d * crossings)
        pi_estimates.append(pi_estimate)
    else:
        pi_estimates.append(0)  # Avoid division by zero early on

# Plotting convergence
plt.figure(figsize=(10, 6))
plt.plot(range(1, N + 1), pi_estimates, label='Estimated π')
plt.axhline(y=np.pi, color='r', linestyle='--', label='True π')
plt.xscale('log')
plt.xlabel('Number of Needle Drops (log scale)')
plt.ylabel('Estimated π Value')
plt.title('Convergence of π Estimate using Buffon\'s Needle (Monte Carlo)')
plt.legend()
plt.grid(True)
plt.show()

# Final output
print(f"Final Estimate of π after {N} drops: {pi_estimates[-1]}")
print(f"True Value of π: {np.pi}")
print(f"Absolute Error: {abs(pi_estimates[-1] - np.pi)}")
```
